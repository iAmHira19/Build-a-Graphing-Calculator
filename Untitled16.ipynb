```python
# Install necessary libraries
!pip install matplotlib numpy sympy pandas

# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sympy as sp

# Function to plot one or more functions
def plot_functions(functions, x_range):
    x = np.linspace(x_range[0], x_range[1], 400)
    plt.figure(figsize=(10, 6))

    for func in functions:
        y = func(x)
        plt.plot(x, y, label=f"{func.__name__}(x)")

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Function Graphs')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example functions
def f1(x):
    return x**2

def f2(x):
    return x**3

# Plot example functions
plot_functions([f1, f2], (-10, 10))

# Function to create a table of (x, y) values
def create_table(function, x_range):
    x = np.linspace(x_range[0], x_range[1], 10)
    y = function(x)
    table = pd.DataFrame({'x': x, 'y': y})
    return table

# Create and display a table for f1
table = create_table(f1, (-10, 10))
print(table)

# Function to shade area above or below the line
def shade_area(function, x_range, shade_above=True):
    x = np.linspace(x_range[0], x_range[1], 400)
    y = function(x)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f"{function.__name__}(x)")

    if shade_above:
        plt.fill_between(x, y, max(y), color='lightblue', alpha=0.5)
    else:
        plt.fill_between(x, y, min(y), color='lightcoral', alpha=0.5)

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Shaded Area')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example of shading area above f1
shade_area(f1, (-10, 10), shade_above=True)

# Function to solve and graph a system of equations
def plot_system_of_equations(eq1, eq2, x_range):
    x = np.linspace(x_range[0], x_range[1], 400)

    # Convert SymPy equations to functions
    x_sym, y_sym = sp.symbols('x y')
    func1 = sp.lambdify(x_sym, sp.solve(eq1, y_sym)[0], 'numpy')
    func2 = sp.lambdify(x_sym, sp.solve(eq2, y_sym)[0], 'numpy')

    plt.figure(figsize=(10, 6))
    plt.plot(x, func1(x), label='Equation 1')
    plt.plot(x, func2(x), label='Equation 2')

    # Solve equations
    solutions = sp.solve((eq1, eq2), (x_sym, y_sym))

    if solutions:
        x_sol = float(solutions[x_sym])
        y_sol = float(solutions[y_sym])
        plt.scatter(x_sol, y_sol, color='red', zorder=5, label='Solution')

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('System of Equations')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example system of equations
x, y = sp.symbols('x y')
eq1 = sp.Eq(x + y, 2)
eq2 = sp.Eq(x - y, 0)
plot_system_of_equations(eq1, eq2, (-10, 10))

# Function to solve quadratic equations
def solve_quadratic(a, b, c):
    x = sp.symbols('x')
    eq = sp.Eq(a*x**2 + b*x + c, 0)
    solutions = sp.solve(eq, x)
    return solutions

# Solve and print solutions for a quadratic equation
solutions = solve_quadratic(1, -3, 2)
print(f"Solutions to the quadratic equation: {solutions}")
```
